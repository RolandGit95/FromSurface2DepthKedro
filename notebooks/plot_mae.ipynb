```python
import numpy as np
import matplotlib.pyplot as plt 
import os, sys, glob, re
```

```python
folder = '../data/05_validation/regimeB'
#prefix = 'STLSTM_t_1_6_12_17_22_27_32_d_'
#prefix = 'STLSTM_t_1_6_12_18_24_32_v2_d_'
#prefix = 'STLSTM_t_1_6_12_18_24_32_v2_d_'
prefix = 'STLSTM_t_1_6_12_18_24_32_v2_d_'
prefix = 'STLSTM_t_0_2_4_6_8_10_12_14_16_18_20_22_24_26_28_30_31_d_'
full = os.path.join(folder, prefix + '*')

def atoi(text):
    return int(text) if text.isdigit() else text

def natural_keys(text):
    return [atoi(c) for c in re.split(r'(\d+)', text)]

files = glob.glob(full)
files.sort(key=natural_keys)

losses = []
for i, file in enumerate(files):
    l = np.load(file)
    losses.append(l)
    
plt.plot(losses)
```
